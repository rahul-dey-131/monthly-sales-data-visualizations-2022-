        Step-1:    Importing Required Modules...    

We'll import the required modules at first. We'll need `Matplotlib` and `Pandas` modules.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

        Step-2:    Loading the Data-set...     
Now, let's import the  `sales_data_2022.csv` using `pandas` module and save it to a variable called `dframe`

```python
dframe = pd.read_csv("sales_data_2022.csv")
dframe.head(13)
```

---

        Step-3:    Analyzing and Plotting...    

(a)    Line plot of Product_A_Sales and Product_B_Sales over 12 months:

```python
# Setting the Figure
plt.style.use('dark_background')
plt.figure(figsize=(13,6))

# Plotting
plt.plot(dframe['Month'], dframe['Laptop'], marker = 'o', label = 'Lapot sold', color = 'cyan', linestyle = 'dotted', linewidth = '2', markersize = 10)
plt.plot(dframe['Month'], dframe['Tab'], marker = 'x', label = 'Tab sold', color = 'orange', linestyle = 'dashed', linewidth = '2', markersize = 10)

# Organizing the Visualization
plt.title("Monthly Sales of Laptop and Tab (2022)")
plt.xlabel('Months')
plt.ylabel('Sales')
plt.xticks(rotation = 45)
plt.legend()
plt.tight_layout()

# Displaying the Plot
plt.show()
```

(b) A bar chart of Total_Sales vs Target_Sales (Highlighted the month(s) where Total_Sales exceeded Target_Sales):


```python
# Setting the Figure
plt.style.use('dark_background')
plt.figure(figsize=(13,6))

# Plotting
x = range(len(dframe))
bar_width = 0.4
colors = ['cyan' if dframe['Total_Sales'][i] >= dframe['Target_Sales'][i] else 'grey' for i in x]      # Apllying conditions to highlight the months where total sales exceed the target sales
plt.bar([i - bar_width / 2 for i in x], dframe['Total_Sales'], width = bar_width, label = 'Total Sales', color = colors)
plt.bar([i + bar_width / 2 for i in x], dframe['Target_Sales'], width = bar_width, label = 'Target Sales', color = 'orange')

# Organizing and Visualization
plt.xticks(ticks = x, labels = dframe['Month'], rotation = 45)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title("Total Sales vs Target Sales (2022)")
plt.legend()
plt.tight_layout()
plt.grid(axis = 'y', linestyle = 'dashed', color = 'grey', alpha = 0.5)
plt.grid(axis = 'x', linestyle = '')

# Displaying the Plot
plt.show()
```
